```python
import numpy as np
import pandas as pd
import plotly.express as px
```

Load the annotated results

Load Salmon-map DE results

```python
salmon_map_DE = pd.read_csv("C:/Users/Sebas/OneDrive/Dokument/Master courses/MASTER THESIS/R project-Master Thesis/salmon_map_dominance_DE_sex_results.csv", float_precision='legacy')
# Replace the zeros in padj with 1e-308 avoid log10 issues
salmon_map_DE["padj_safe"] = salmon_map_DE["padj"].replace(0, 1e-308).fillna(1)

# Add the negative log 10 padj for plotting
salmon_map_DE["neglog10_padj"] = -np.log10(salmon_map_DE["padj_safe"])


# Add significance to differentially expressed transcripts 
salmon_map_DE["significant"] = (
    (salmon_map_DE["padj"] < 0.05) &
    (salmon_map_DE["log2FoldChange"].abs() > 1)
)

# Add a label to the significant transcripts
salmon_map_DE["label"] = salmon_map_DE["transcript_id"].where(salmon_map_DE["significant"], "")

salmon_map_DE
```
